```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.read_gml("../subway_graph.gml")
print(f"Graph has {G.number_of_nodes()} stations and {G.number_of_edges()} connections.")
```

```python
plt.figure(figsize=(30, 24))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_size=400, font_size=8)
plt.show()
```

```python
G_directed =  nx.DiGraph(G)
print(f"Graph has {G_directed.number_of_nodes()} stations and {G_directed.number_of_edges()} connections.")
```

```python
plt.figure(figsize=(30, 24))
pos = nx.spring_layout(G_directed, seed=42)
nx.draw(G, pos, with_labels=True, node_size=400, font_size=8)
plt.show()
```
